#### Splitting data set 

```python
x_values = data_set.drop(['chocolate'],axis=1).values
y_values =data_set['chocolate'].values
```

```python
sc_x = StandardScaler()
x_values = sc_x.fit_transform(x_values)
```

```python
#obtaining splitted training and test data set
train_X, test_X, train_Y, test_Y = model_selection.train_test_split(x_values,y_values,test_size=0.2,random_state=0)
```

#### Obtaining Classification Model 

```python
##  USING LOGICTIC REGRESSION FOR THIS CLASSIFICATION
classifier = fit_or_predict(train_X,train_Y,LogisticRegression(),1)
```

#### Predicting Output

```python
test_prediction = fit_or_predict(test_X,test_Y,classifier)

train_prediction = fit_or_predict(train_X,train_Y,classifier)
```

#### Checking Confusion Matrix and score

```python
print("Confusion Matrix For Test data Set  :  ","\n",confusion_matrix(test_prediction,test_Y))
print("Confusion Matrix For Train data Set :  ","\n",confusion_matrix(train_prediction,train_Y))
```

```python
print("Accuracy Score For Test data Set  :  ","\n",accuracy_score(test_prediction,test_Y)*100)
print("Accuracy Score For Train data Set :  ","\n",accuracy_score(train_prediction,train_Y)*100)
```
