Problem Statement :

Generate association rules for dataset given in the url
https://drive.google.com/open?id=1ke8crGAhKwSx1xbanwr8n5gNwpqOdDXU


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

##### Obtaining Data and Making A List Of Lists  >>

```python
data_set = open('/home/admin3/Documents/MyDoc/data_sets/Market_Basket_Optimisation.csv').read().splitlines()
transactions = [lines.split(",") for lines in data_set]
```

```python
transactions[:5]
```

##### Obtaining Rules 

```python
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

##### Visualising Results

```python
results = list(rules)
```

```python
for detail in results:
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
    print("Items Bought Together >>>>>>>",detail.items)
    print("Support    for items  >>>>>>>",detail.support)
    print()
    print(detail.ordered_statistics)
    print()
    print()
```
